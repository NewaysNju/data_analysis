```python
# matplotlib图像都位于Figure对象中，用plt.figure创建新的Figure
fig = plt.figure()
# 不能通过空Figure绘图，必须用add_subplot创建一个或多个subplot才行
ax1 = fig.add_subplot(2, 2, 1)  # 这里是创建2x2个图像，选中第一个
ax2 = fig.add_subplot(2, 2, 2)  
ax3 = fig.add_subplot(2, 2, 3)  # 此时的当前画板是第三个,返回的对象是AxesSubplot对象
from numpy.random import randn
plt.plot(randn(50).cumsum(), 'k--')  
```

```python
# 注解  --- text, arrow, annotate
from datetime import datetime
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
data = pd.read_csv(r'spx.csv',index_col = 0, parse_dates = True)
spx = data['SPX']
crisis_data = [
    (datetime(2007, 10, 11), 'Peak of bull market'),
    (datetime(2008, 3, 12), 'Bear Stearns Fails' ),
    (datetime(2008, 9, 15), 'Lehman Bankruptcy')
]
for date, label in crisis_data:
    ax.annotate(label, xy = (date, spx.asof(date) + 50),  # Series.asof(index),根据index获取value == obj.['index']
               xytext = (date, spx.asof(date) + 200),
               arrowprops = dict(facecolor = 'black'),
               horizontalalignment = 'left', verticalalignment = 'top')
# 放大到2007-2010
ax.set_xlim(['1/1/2007', '1/1/2011'])
ax.set_ylim([600, 1800])
ax.set_title('Important dates in 2008-2009 financial crisis')
```

```python
# 添加图形
# 先创建块对象shp，然后通过ax.add_patch(shp)来添加
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
rect = plt.Rectangle((0.2, 0.75), 0.4, 0.15, color = 'k', alpha = 0.3)  # 左下角坐标，然后是长和宽
circ = plt.Circle((0.7, 0.2), 0.15, color = 'b', alpha = 0.3)  # 原点坐标 ， 半径
pgon = plt.Polygon([[0.15, 0.15], [0.35, 0.4], [0.2, 0.6]], color = 'g', alpha = 0.5)  # 三个点坐标
ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(pgon)
```

## 用于Series的参数  
* label:图例标签  
* ax:画布选择  
* style:风格字符串， 如ko--  
* alpha:透明度  
* kind:形状种类   
* logy:y轴使用对数标尺  
* use_index:将对象的索引用用作刻度标签  
* xticks/yticks：刻度值<br>
* xlim/ylim<br>
* grid:打格子<br>

```python
data = data[(data.LATITUDE>18)&(data.LATITUDE<20)&
           (data.LONGITUDE>-75)&(data.LONGITUDE<-70)
           &data.CATEGORY.notnull()]  # 筛选
```

```python
def to_cat_list(catstr): 
    # 将字符串按,分开，转成一个list
    stripped = (x.strip() for x in catstr.split(','))
    return [x for x in stripped if x]
def get_all_categories(cat_series):
    # 获取类
    cat_sets = (set(to_cat_list(x)) for x in cat_series)
    return sorted(set.union(*cat_sets))
def get_english(cat):
    # 通过分开获取代码和英文名
    code, names = cat.split('.')
    if '|' in names:
        names = names.split(' | ')[1]
    return code, names.strip()
```

```python
get_english('2. Urgences logistiques | Vital Lines')
```

```python
all_cats = get_all_categories(data.CATEGORY)
english_mapping = dict(get_english(x) for x in all_cats)
english_mapping['2a']
```

```python
english_mapping['6c']
```

```python
# 构造哑变量 -- 构造全0矩阵
def get_code(seq):
    return [x.split('.')[0] for x in seq if x]
all_codes = get_code(all_cats)
code_index = pd.Index(np.unique(all_codes))
dummy_frame = DataFrame(np.zeros((len(data), len(code_index))), index = data.index,
                       columns = code_index)
dummy_frame.iloc[:10, :6]
```
